```python
import pandas as pd
import numpy as np
import pandasql as ps
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv("global_food_wastage_dataset_with_coords.csv")
```

```python
df.head()
```

```python
fig = px.scatter_geo(
    df,
    lat="latitude",
    lon="longitude",
    color="Total Waste (Tons)",
    size="Total Waste (Tons)",
    animation_frame="Year",
    hover_name="Country",
    projection="natural earth",
    title="Global Food Waste Heat Map Over the Years",
    color_continuous_scale="Reds"
)
```

```python
fig.update_layout(
    width=1200,
    height=700,
    geo=dict(showland=True, landcolor="white", showcountries=True),
    title_x=0.5,
    margin=dict(l=20, r=20, t=50, b=20),
    legend_title="Total Waste (Tons)"
    )
```

```python
fig = px.scatter_geo(
    df,
    lat="latitude",
    lon="longitude",
    hover_name="Country",
    hover_data=["Food Category", "Total Waste (Tons)", "Economic Loss (Million $)"],
    size="Total Waste (Tons)",
    color="Food Category",
    projection="natural earth",
    title="Global Food Waste Overview"
)
```

```python
fig.update_layout(
    width=1200,
    height=700,
    title_font_size=24,
    margin={"r":20, "t":50, "l":20, "b":20}
)

fig.show()
```
